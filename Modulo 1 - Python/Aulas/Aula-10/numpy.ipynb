**Lista e Array**

```python
list = [10, 20, 30]
array = np.array([list])

print(list + [5]) # append the number 5 to the list
print(array + 5) # sum 5 to each item of the list - "vetorização"
```

**Vetorização**

```python
prices = [100, 200, 300]
discounted_prices = []

for price in prices:
    discounted_prices.append(price * 0.9)

print(discounted_prices)
```

```python
prices = np.array([100, 200, 300]) # It´s a python's list -> array
discounted_prices = prices * 0.9

print(discounted_prices)
```

```python
grades = np.array([
    [8.0, 10, 8.5],
    [9.2, 7.5, 4.5],
    [10, 8.9, 5.5],
    [4.5, 8.6, 9]
])

print(grades.ndim) # dimension
print(grades.shape) # number of rows and columns
print(grades.size) # total number of elements
print(grades.dtype)
```

```python
grades = np.array([
    [8.0, 10, 8.5],
    [9.2, 7.5, 4.5],
    [10, 8.9, 5.5],
    [4.5, 8.6, 9]
])

print(grades[0, 0])
print(grades[0])
print(grades[0, 0], grades[3, 2])
print(grades[-1, -1])
print(grades[:, 0]) # Column
```

```python
grades = np.array([8.0, 10, 8.5])

print(grades + 1)
print(grades - 1)
print(grades * 1)
print(grades / 2)

```

**Operações entre Arrays**

```python
grades_av1 = np.array([8.0, 10, 8.5, 10])
grades_av2 = np.array([4.5, 8, 7.5, 9])

average_grades = ((grades_av1 + grades_av2) / 2)
print(average_grades)
```

```python
grades_av1 = np.array([8.0, 10, 8.5, 10])

print(grades_av1.sum())
print(grades_av1.mean())
print(grades_av1.min())
print(grades_av1.max())
print(grades_av1.std())
```

```python
grades_first_semester = np.array([8.0, 10, 8.5, 10])
grades_second_semester = np.array([4.5, 8, 7.5, 9])

average = np.mean((grades_first_semester + grades_second_semester) / 2)

print(average)  # Overall average grade
```

**Média por coluna**

```python
grades = np.array([
    [8.0, 10, 8.5],
    [9.2, 7.5, 4.5],
    [10, 8.9, 5.5],
    [4.5, 8.6, 9]
])

print(grades.mean(axis=0))
```

```python
grades = np.array([8.0, 10, 8.5, 10])

print(grades > 8)
print(grades[grades > 8])
print((grades > 8).sum())
```

```python
grades= np.array([8.0, 10, 8.5, 10, 5.5])


average = grades.mean()
aprovados = average >= 7
print(grades[aprovados])
print(grades[aprovados])
```
